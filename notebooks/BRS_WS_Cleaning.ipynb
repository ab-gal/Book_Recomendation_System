```python
#dropping columns
df_reduced = df.drop(
    columns=[
        "edition_count",
        "pages",
        "book_url",
        "publisher"
    ]
)
```

```python
for idx, author in author_updates.items():
    df_reduced.loc[idx, "author"] = author
```

```python
language_updates = {
    76: "English, French",
    101: "English",
    217: "English",
    218: "English",
    273: "English",
    285: "German",
    286: "English",
    290: "English",
    359: "English",
    362: "German",
    365: "English",
    370: "English",
    426: "French",
    429: "English",
    430: "English",
    445: "English",
    447: "English",
    484: "English",
}
for idx, language in language_updates.items():
    df_reduced.loc[idx, "language"] = language
```

```python
df_reduced.loc[list(language_updates.keys()), ["title", "language"]]
```

```python
df_reduced.info()
```

```python
#Cleaning Text Columns
text_columns = ["title", "author", "language", "subjects"]

for col in text_columns:
    df_reduced[col] = (
        df_reduced[col]
        .fillna("")
        .str.lower()
        .str.replace("\n", " ", regex=False)
        .str.replace("\r", " ", regex=False)
        .str.strip()
    )
text_columns
```

```python
df_reduced.to_csv("data/raw/web_scrapping_no_null.csv", index=False)
```
